# In addition, you will be expected to produce the following three pie charts:

We created 3 different values called 1. Urban 2. Suburban 3. Rural 
Values were made by using groupby function and using sum, count, mean functions

After creating values created scatter plot step by step  implementing Urban, Suburban and Rural accordingly
After creating the Scatter plot , we added edging to the circles with alpha.

we created 3 pie charts based on the requirement  Total Fares city type, Total rides by city type and Total Rides by city types

```python
#Urban conditiona statemnt is ==
urban_cities = pyber_table.loc[pyber_table["type"] =="Urban"]

suburban_cities = pyber_table.loc[pyber_table["type"] =="Suburban"]

rural_cities = pyber_table.loc[pyber_table["type"] =="Rural"]


##groupby on urban
#urban ride count
urban_data = urban_cities.groupby("city")["ride_id"].count()

# urban drivercount
urban_dr_count = urban_cities.groupby("city")["driver_count"].sum()

#urban average-FARE
urban_avg_fare = urban_cities.groupby("city")["fare"].mean()
urban_avg_fare.round(2)

#suburban_data = suburban_cities("")
suburban_data = suburban_cities.groupby("city")["ride_id"].count()

suburban_dr_count = suburban_cities.groupby("city")["driver_count"].sum()
suburban_avg_fare = suburban_cities.groupby("city")["fare"].mean()
suburban_avg_fare.round(2)

## rural cities
rural_data = rural_cities.groupby("city")["ride_id"].count()
rural_dr_count = rural_cities.groupby("city")["driver_count"].sum()
rural_avg_fare = rural_cities.groupby("city")["fare"].mean()
rural_avg_fare.round(2)
```

 ## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

# Build the scatter plots for each city types

# Incorporate the other graph properties

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
```

```python
## creating scatter plot by assigning values

  

plt.scatter(urban_data, urban_avg_fare, s=urban_dr_count*.4, c="coral", alpha=.85,edgecolor="black",linewidths=1)
plt.scatter(suburban_data, suburban_avg_fare, s=suburban_dr_count*.9, c="lightskyblue", alpha=1,edgecolor="black",linewidths=1)
plt.scatter(rural_data, rural_avg_fare, s=rural_dr_count, c="gold", alpha=1,edgecolor="black",linewidths=1)
plt.title("Pyber Ride sharing Data (2016)")
plt.xlabel("Total Number of Rides Per City")
plt.ylabel("Average Fare ($) Per City")
plt.legend(('Urban','Suburban','Rural'),loc = "upper right",title = "City Type",markerscale=.4)
plt.text(45,35,"Notes:\nCircle size correlates with driver count per City")
plt.grid(True)

plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents

# calculate variables that will be used in the Pie Chart


# Calculate Total fare($) Per CityType
ctfare = pyber_table.groupby(["type"])["fare"].sum()
ctfare
ctfare.index


# Calculate Total Rides Per CityType
ctride = pyber_table.groupby(["type"])["ride_id"].count()

# Calculate Total Drivers Per CityType
ctdrive = pyber_table.groupby(["type"])["driver_count"].sum()
```

```python
# Build Pie Chart
#Set the colors of each section of the pie chart.
colors = ["gold", "lightskyblue", "lightcoral"]
#explodes the third chart part of the pie
explode = (0,0,0.1)
plt.pie(ctfare, explode=explode, labels=ctfare.index, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
# Save Figure
plt.axis("equal")
plt.title("% of Total Fares by City Type")
#Add the legend.
plt.legend(loc = "upper right",title = "City Type")
# Show Figure
plt.show()
```
